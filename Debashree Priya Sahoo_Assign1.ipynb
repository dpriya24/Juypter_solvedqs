1. What is the Number of Directors in the given dataset ? 

```python
mydata['Director'] = mydata['Director'].str.strip()
num_directors = mydata['Director'].nunique()
print("Number of directors:", num_directors)
```

2. What is the mean, median and standard deviation of IMDB rating in the dataset ?

```python

mean_rating = mydata['IMDB_Rating'].mean()
median_rating = mydata['IMDB_Rating'].median()
std_dev_rating = mydata['IMDB_Rating'].std()
print("Mean rating:", mean_rating)
print("Median rating:", median_rating)
print("Standard deviation of rating:", std_dev_rating)
```

3. Which director has directed the most number of movies?

```python
mydata['Director'] = mydata['Director'].str.strip()
director_counts = mydata['Director'].value_counts()
most_movies_director = director_counts.idxmax()
print("Director with the most number of movies:", most_movies_director)
```


4. Name the Movie which has the highest Runtime ?

```python
mydata=pd.read_csv("C:/Users/debas/OneDrive/Desktop/imdb_top_1000.csv")
mydata['Runtime'] = pd.to_numeric(mydata['Runtime'], errors='coerce')
valid_runtime_movies = mydata.dropna(subset=['Runtime'])
max_runtime_movie = valid_runtime_movies.loc[valid_runtime_movies['Runtime'].idxmax(), 'Series_Title']
print("Movie with the highest runtime:", max_runtime_movie)
```

5. How many movies are released after year 2000 and have IMDB rating > 8.5 ?

```python

mydata['Released_Year'] = pd.to_numeric(mydata['Released_Year'], errors='coerce')
mydata['IMDB_Rating'] = pd.to_numeric(mydata['IMDB_Rating'], errors='coerce')
filtered_mydata = mydata[(mydata['Released_Year'] > 2000) & (mydata['IMDB_Rating'] > 8.5)]
num_movies = len(filtered_mydata)
print("Number of movies released after 2000 with IMDb rating > 8.5:", num_movies)
```

8. What type of Genre has the highest IMDB rating?

```python

genre_ratings = mydata.groupby('Genre')['IMDB_Rating'].mean()

highest_rating_genre = genre_ratings.idxmax()


print("Genre with the highest IMDb rating:", highest_rating_genre)

```

9. Name the top 5 movies that have earned highest Meta score. 

```python
mydata=pd.read_csv("C:/Users/debas/OneDrive/Desktop/imdb_top_1000.csv")
sorted_data = mydata.sort_values('Meta_score', ascending=False)
top_movies = sorted_data.head(5)['Series_Title']
print("Top 5 movies with the highest Meta score:")
for movie in top_movies:
    print(movie)
```
